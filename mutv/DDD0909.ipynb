## Year by year batting

```python
yxy_batting = pd.read_csv('https://raw.githubusercontent.com/gcdean38/MSM_python/main/mutv/DDD0909_1.csv')
```

```python
yxy_batting = yxy_batting[['Year','R/G','G','PA','AB','R','H','HR','BB','SO','BA','OBP','SLG','OPS','TB']]
```

```python
yxy_batting.dtypes
```

```python

yxy_batting['decade'] = yxy_batting['Year'].astype(str).str[:3]+"0s"
yxy_batting
```

```python
yxy_batting['H_HR_pct'] = yxy_batting['HR'] / yxy_batting['H']
yxy_batting['SO_pct'] = yxy_batting['SO'] / yxy_batting['PA']
yxy_batting['H_pct'] = yxy_batting['H'] / yxy_batting['PA']
yxy_batting['BB_pct'] = yxy_batting['BB'] / yxy_batting['PA']
yxy_batting['HR_pct'] = yxy_batting['HR'] / yxy_batting['PA']
```

```python
yxy_batting.head(15)
```

```python
yxy_batting.plot.line(x="Year",y="H_HR_pct")
```

```python
yxy_batting.plot.line(x="Year",y=['SO_pct','BB_pct'])
```

```python
yxy_batting.plot.line(x="Year",y=['HR_pct'])
```

```python
yxy_batting.plot.line(x="Year",y=['SO_pct'])
```

```python
r = np.corrcoef(yxy_batting['SO_pct'], yxy_batting['HR_pct'])
r
```

```python
dxd_batting = yxy_batting.groupby('decade')[['R/G','H','HR','SO','BB','PA']].mean()
dxd_batting
```

```python
dxd_batting['H_HR_pct'] = dxd_batting['HR'] / dxd_batting['H']
dxd_batting['SO_pct'] = dxd_batting['SO'] / dxd_batting['PA']
dxd_batting['H_pct'] = dxd_batting['H'] / dxd_batting['PA']
dxd_batting['BB_pct'] = dxd_batting['BB'] / dxd_batting['PA']
dxd_batting['HR_pct'] = dxd_batting['HR'] / dxd_batting['PA']
```
