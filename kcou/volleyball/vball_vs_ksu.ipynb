```python
import pandas as pd
```

```python
sets = pd.read_csv('https://raw.githubusercontent.com/gcdean38/MSM_python/main/kcou/volleyball/miz_vb_2020_set_res.csv')
```

```python
sets.head()
```

```python
sets['set_count'] = sets['miz_sets'].astype(str) + "-" + sets['opp_sets'].astype(str)
```

```python
sets.head()
```

## Set win prob

```python
set_win_prob = pd.crosstab(index = sets['set_count'], columns = sets['set_res'])
```

```python
set_win_prob['win_pct'] = set_win_prob['W'] / (set_win_prob['W']+set_win_prob['L'])
```

```python
set_win_prob
```

## Match win prob

```python
match_win_prob = pd.crosstab(index = sets['set_count'], columns = sets['match_res'])
```

```python
match_win_prob['win_pct'] = match_win_prob['W'] / (match_win_prob['W']+match_win_prob['L'])
```

```python
match_win_prob
```
