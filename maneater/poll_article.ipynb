## All seasons

```python
import pandas as pd
```

```python
base_url = "http://collegepollarchive.com/football/ap/seasons.cfm?seasonid="
pd.read_html(base_url)
urls = []
for year in range(1990,2021):
    url = base_url + str(year)
    urls.append(url)
print(urls)
```

```python
stayed_list = []
same_list = []
print("Teams that were in the AP Top Ten preseason and final poll")
for url in urls:
    tables = pd.read_html(url)
    table = tables[9]
    
    poll_table = table[['Rank',"Rank.1","Team (FPV)"]]
    poll_table = poll_table[:11]
    
    # format table
    poll_table['final_rank'] = poll_table['Rank']
    poll_table[['dummy','preseason_rank']] = poll_table['Rank.1'].str.split('<',expand=True)
    
    # format table cont.
    poll_table = poll_table[['Team (FPV)','final_rank','preseason_rank']]
    poll_table['final_rank'] = poll_table['final_rank'].astype(int)
    poll_table['preseason_rank'] = poll_table['preseason_rank'].astype(int)
    
    # Stayed in Top Ten Column
    poll_table.loc[poll_table['preseason_rank'] <= 10, 'top_ten'] = 1
    poll_table.loc[poll_table['preseason_rank'] > 10, 'top_ten'] = 0
    total_stayed = sum(poll_table['top_ten'])
    stayed_list.append(total_stayed)
    
    # Same rank column
    poll_table.loc[poll_table['preseason_rank'] == poll_table['final_rank'], 'same'] = 1
    poll_table.loc[poll_table['preseason_rank'] != poll_table['final_rank'], 'same'] = 0
    total_same = sum(poll_table['same'])
    same_list.append(total_same)
    
    print(url[-4:],"Stayed in top ten: ", sum(poll_table['top_ten']), "| Kept same ranking: ", sum(poll_table['same']))
    
```

```python
r = np.corrcoef(stayed_list, same_list)
r
```
